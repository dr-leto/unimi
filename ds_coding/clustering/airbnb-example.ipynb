## Create a map of accommodations with price
- bokeh
- pyplot
- seaborn
- plotly

```python
import matplotlib.pyplot as plt
```

```python
def plot_prices(dfs, ax, color_variable, boxplots_variables, cmap='Reds'):
    if len(ax) < 1 + len(boxplots_variables):
        raise Exception('The number of axes should be higher than number of variables')
    for df in dfs:
        ax[0].scatter(df.longitude, df.latitude, c=df[color_variable], 
                      alpha=0.4, cmap=cmap)
        for i, var in enumerate(boxplots_variables):
            df.boxplot(column=[var], ax=ax[i+1])
    plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(14, 4), ncols=3)
plot_prices([milan], ax, color_variable='daily_price', 
            boxplots_variables=['daily_price', 'review_scores_rating'])
plt.show()
```

### Ideas for handling bad distribution of daily price
- remove outliers
- scale (log, scaling in range, kernel)
- discretize (clustering)

#### Using a threshold

```python
th = 300
lp = milan[milan.daily_price <= th]
hp = milan[milan.daily_price > th]
```

```python
fig, ax = plt.subplots(figsize=(14, 4), ncols=3)
plot_prices([lp, hp], ax, color_variable='daily_price', 
            boxplots_variables=['daily_price', 'review_scores_rating'])
plt.show()
```

```python
th = np.percentile(milan.daily_price, [98])[0]
lp = milan[milan.daily_price <= th]
hp = milan[milan.daily_price > th]
```

```python
fig, ax = plt.subplots(figsize=(14, 4), ncols=3)
plot_prices([lp, hp], ax, color_variable='daily_price', 
            boxplots_variables=['daily_price', 'review_scores_rating'])
plt.show()
```

#### Scale

```python
milan['log_price'] = np.log(milan.daily_price)
```

```python
fig, ax = plt.subplots(figsize=(14, 4), ncols=3)
plot_prices([milan], ax, color_variable='log_price', 
            boxplots_variables=['daily_price', 'log_price'])
plt.show()
```

#### Sklearn scalers

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scalers = [StandardScaler(), MinMaxScaler(feature_range=(0, 1))]
data = milan.daily_price.values.reshape(-1, 1)
for scaler in scalers:
    scaler.fit(data)
    data = scaler.transform(data)
```

#### Discretize

```python
thresholds = np.percentile(milan.daily_price, [25, 50, 75])
thresholds
```

```python
clusters = []
for price in milan.daily_price.values:
    cluster = len(thresholds)
    for i, th in enumerate(thresholds):
        if price <= th:
            cluster = i
    clusters.append(cluster)
```

```python
milan.daily_price.values[:4]
```

```python
clusters[:4]
```

```python
milan['perc_clusters'] = clusters
```

```python
fig, ax = plt.subplots(figsize=(14, 4), ncols=3)
plot_prices([milan], ax, color_variable='perc_clusters', 
            boxplots_variables=['daily_price', 'perc_clusters'], cmap='Accent')
plt.show()
```
