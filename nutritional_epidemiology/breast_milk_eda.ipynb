## PCA with nutrients from dietary habits

```python
import re

# Select columns corresponding to the dietary habits
raw_nutrient_cols = np.array([c for c in data.columns if re.match(r"nut[0-9]+", c)])

# Drop columns with constant values
nutrient_cols = raw_nutrient_cols[data[raw_nutrient_cols].nunique() != 1]

print("Dropped %d columns with constant values" % (len(raw_nutrient_cols) - len(nutrient_cols)))
raw_nutrients = data[nutrient_cols].values
```

```python
# Standartize the data
nutrients = (raw_nutrients - raw_nutrients.mean(axis=0)) / raw_nutrients.std(axis=0)
```

```python
from numpy.linalg import svd

# Apply SVD decomposition
u, s_arr, v_t = svd(nutrients, full_matrices=False)
s = np.zeros(shape=(len(s_arr), len(s_arr)))
```

```python
# Explained variance depending on the number of components selected

explained_variance = []
ks = list(range(1, 20))
for k in ks:
    explained_variance.append(np.sum(s_arr[:k]**2) / np.sum(s_arr**2))

sns.lineplot(x=ks, y=explained_variance)
plt.xticks(ks)
plt.title("Explained variance by number of components")
plt.xlabel("K")
plt.ylabel("Explained variance ratio");
```

```python
best_k = len(s_arr)
np.fill_diagonal(s, s_arr[:best_k])
loadings = v_t.T @ s / np.sqrt(nutrients.shape[0])
```

```python
for k in range(5):
    col_weights = zip(nutrient_cols, loadings[k, :])
    col_weights = sorted(col_weights, key=lambda x: -x[1])
    features = [c[0] for c in col_weights]
    weights = [c[1] for c in col_weights]

    n_features = 10
    sns.barplot(x=features[:n_features], y=weights[:n_features]);
    plt.xlabel("Factors")
    plt.ylabel("PCA loadings")
    plt.title(f"Factor loadings for {k + 1} component")
    plt.show()
```
