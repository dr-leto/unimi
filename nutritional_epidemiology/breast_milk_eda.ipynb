## PCA with nutrients from dietary habits

```python
import re

# Select columns corresponding to the dietary habits
# raw_nutrient_cols = np.array([c for c in data.columns if re.match(r"nut[0-9]+", c)])

# Drop columns with constant values
# nutrient_cols = raw_nutrient_cols[data[raw_nutrient_cols].nunique() != 1]

# print("Dropped %d columns with constant values" % (len(raw_nutrient_cols) - len(nutrient_cols)))

diet_nutrient_cols = ["nut2", "nut3", "nut9", "nut10", "nut11", "nut14", "nut15", "nut16",
             "nut17", "nut18", "nut19", "nut20", "nut21", "nut22", "nut23", "nut25", "nut26",
             "nut29", "nut30", "nut31", "nut32", "nut33", "nut34", "nut35", "nut36", 
             "nut37", "nut46", "nut92", "nut93", "nut96", "nut97"]

raw_diet_nutrients = data[diet_nutrient_cols].values
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

sc = StandardScaler()
diet_nutrients = sc.fit_transform(raw_diet_nutrients)

pca = PCA(n_components=None)
pca_model = pca.fit(diet_nutrients)
```

```python
sns.lineplot(x=list(range(1, raw_diet_nutrients.shape[1] + 1)), y=np.cumsum(pca_model.explained_variance_ratio_))
plt.title("Explained variance by number of components")
plt.xlabel("K")
plt.ylabel("Explained variance ratio");
```

```python
pca_loadings = pca_model.components_.T @ np.diag(np.sqrt(pca_model.singular_values_))
```

```python
for k in range(5):
    col_weights = zip(diet_nutrient_cols, pca_loadings[:, k])
    col_weights = sorted(col_weights, key=lambda x: -x[1])
    features = [c[0] for c in col_weights]
    weights = [c[1] for c in col_weights]

    n_features = 10
    sns.barplot(x=features[:n_features], y=weights[:n_features]);
    plt.xlabel("Factors")
    plt.ylabel("PCA loadings")
    plt.title(f"Factor loadings for {k + 1} component")
    plt.show()
```

## Correlation of dietary patterns with maternal milk composion

```python
milk_nutrient_cols = ["nutm4", "nutm5", "nutm6", "nutm8", "nutm50", "nutm54", "nutm64", "nutm67", "nutm75",
  "nutm77", "nutm80", "nutm81", "nutm82", "nutm83", "nutm85", "nutm87"]
milk_nutrients_names = ["Fat", "Protein" , "Lactose", "Energy", "LA" , "ALA", "AA" , "EPA" , "DPA",
                        "dha", "omega 3", "Omega 6", "PUFA", "omega 6/ omega 3 ratio" , "SFA", "mufa"]

milk_nutrients_mapping = dict(zip(milk_nutrient_cols, milk_nutrients_names))

raw_milk_nutrients = data[milk_nutrient_cols].values
```

```python
k = 5
dietary_patterns = PCA(n_components=k).fit_transform(diet_nutrients)
```

```python
from numpy import corrcoef

np_dietary_vs_milk_corr = np.zeros(shape=(len(milk_nutrient_cols), k))
for i in range(len(milk_nutrient_cols)):
    for j in range(k):
        np_dietary_vs_milk_corr[i][j] = corrcoef(raw_milk_nutrients[:, i], dietary_patterns[:, j])[0][1]
        
dietary_pattern_names = [f"dietary_pattern_{i}" for i in range(k)]
df_dietary_vs_milk_corr = pd.DataFrame(data=np_dietary_vs_milk_corr, columns=dietary_pattern_names, index=milk_nutrient_cols)
```

## Using in-built factor analysis

```python
from sklearn.decomposition import FactorAnalysis

fa = FactorAnalysis(rotation="varimax")
dietary_patterns = fa.fit_transform(diet_nutrients)
fa_loadings = fa.components_.T
```

```python
for k in range(5):
    col_weights = zip(diet_nutrient_cols, fa_loadings[:, k])
    col_weights = sorted(col_weights, key=lambda x: -x[1])
    features = [c[0] for c in col_weights]
    weights = [c[1] for c in col_weights]

    n_features = 10
    sns.barplot(x=features[:n_features], y=weights[:n_features]);
    plt.xlabel("Factors")
    plt.ylabel("PCA loadings")
    plt.title(f"Factor loadings for {k + 1} component")
    plt.show()
```

## Clustering of maternal milk based on the features

```python
sc = StandardScaler()
milk_nutrients = sc.fit_transform(raw_milk_nutrients)
reduced_milk_nutrients = PCA(n_components=2).fit_transform(milk_nutrients)

sns.scatterplot(x=reduced_milk_nutrients[:, 0], y=reduced_milk_nutrients[:, 1]);
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
plt.title("Maternal milk visualized with reduced dimension");
```

```python
from sklearn.cluster import KMeans

ks = list(range(1, 20))
ssds = []

for k in ks:
    kmeans = KMeans(n_clusters=k)
    kmeans_model = kmeans.fit(milk_nutrients)
    ssds.append(kmeans_model.inertia_)
    
sns.lineplot(x=ks, y=ssds)
plt.xlabel("K")
plt.ylabel("Sum of squared distances")
plt.title("Tuning K in KMeans")
```

```python
kmeans = KMeans(n_clusters=3, random_state=228)
data["milk_composion_group"] = kmeans.fit_predict(milk_nutrients)

milk_clusters = data[milk_nutrient_cols + ["milk_composion_group"]].groupby("milk_composion_group").mean() \
    .rename(columns=milk_nutrients_mapping)
milk_nutrients_mean = data[milk_nutrient_cols].rename(columns=milk_nutrients_mapping).mean()
```

```python
data[milk_nutrient_cols + ["milk_composion_group"]].groupby("milk_composion_group").size()
```

```python
milk_clusters_mean_deviation = (milk_clusters - milk_nutrients_mean) / milk_nutrients_mean
```

```python
milk_clusters
```

```python
milk_clusters_mean_deviation.iloc[0].sort_values(key=lambda x: -abs(x))
```

- **1 cluster:** high omega 6/ omega 3 ratio, low EPA, high LA, low DHA, **high fat**, high PUFA, low DPA
- **2 cluster:** extremely high EPA, extremely high DPA, extremely high DHA, low omega 6 / omega 3 ratio, high ALA, **low fat**
- **3 cluster:** low EPA, low DHA, low LA, low PUFA, low DPA, low omeg 6 / omega 3 ratio

## Dietary patterns vs milk composure relationship

```python
from sklearn.model_selection import train_test_split

pca_factors = pd.read_csv("pca_factors.csv", sep=";")
X_train, X_test, y_train, y_test = train_test_split(pca_factors.values, 
                                                    data["milk_composion_group"], 
                                                    stratify=data["milk_composion_group"], 
                                                    random_state=228)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=float("inf"), class_weight="balanced", multi_class="ovr", random_state=228)
lr_model = lr.fit(X_train, y_train)

y_train_pred = lr_model.predict(X_train)
y_test_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print("Train report")
print(classification_report(y_train_pred, y_train))

print()
print("Test report")
print(classification_report(y_test_pred, y_test))
```

```python
for class_label in [0, 1, 2]:
    feature_weights = list(zip(pca_factors.columns, lr_model.coef_[class_label]))
    feature_weights = sorted(feature_weights, key=lambda x: -abs(x[1]))
    features, weights = [list(t) for t in zip(*feature_weights)]

    plt.title(f"Feature weights for class {class_label + 1}")
    plt.xlabel("Principal components")
    plt.ylabel("Logistic regression weights")
    plt.xticks(rotation=45)
    sns.barplot(x=features, y=weights)
    plt.show();
```

## Adjusted patterns

```python
pca_factors = pd.read_csv("pca_factors.csv", sep=";")

add_factors_map = {"v4": "num_children",
                   "v5": "maternal_age",
                   "v6": "weight",
                   "v9": "gest_week",
                   "v10": "inf_weight",
                   "v11": "inf_length"}
add_factors = data[add_factors_map.keys()].rename(columns=add_factors_map)
all_factors = pca_factors.join(add_factors)
```

```python
X_train, X_test, y_train, y_test = train_test_split(all_factors.values, 
                                                    data["milk_composion_group"], 
                                                    stratify=data["milk_composion_group"], 
                                                    random_state=228)
```

```python
lr = LogisticRegression(C=1.0, class_weight="balanced", multi_class="ovr", random_state=228, max_iter=1000)
lr_model = lr.fit(X_train, y_train)

y_train_pred = lr_model.predict(X_train)
y_test_pred = lr_model.predict(X_test)
```

```python
print("Train report")
print(classification_report(y_train_pred, y_train))

print()
print("Test report")
print(classification_report(y_test_pred, y_test))
```

```python
for class_label in [0, 1, 2]:
    feature_weights = list(zip(all_factors.columns, lr_model.coef_[class_label]))
    feature_weights = sorted(feature_weights, key=lambda x: -abs(x[1]))
    features, weights = [list(t) for t in zip(*feature_weights)]

    plt.title(f"Feature weights for class {class_label + 1}")
    plt.xlabel("Principal components")
    plt.ylabel("Logistic regression weights")
    plt.xticks(rotation=45)
    sns.barplot(x=weights, y=features, orient="h")
    plt.show();
```
